```python
import numpy as np
import matplotlib.pyplot as plt

# Q1: Operations on the matrix gfg
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')

# i. Sum of all elements
total_sum = np.sum(gfg)
print("Sum of all elements:", total_sum)

# ii. Sum of all elements row-wise
row_wise_sum = np.sum(gfg, axis=1)
print("Sum of all elements row-wise:\n", row_wise_sum)

# iii. Sum of all elements column-wise
column_wise_sum = np.sum(gfg, axis=0)
print("Sum of all elements column-wise:\n", column_wise_sum)
```

```python
# Part (a): Operations on array
array_a = np.array([10, 52, 62, 16, 16, 54, 453])

# i. Sorted array
sorted_array = np.sort(array_a)
print("\nSorted array:", sorted_array)

# ii. Indices of sorted array
sorted_indices = np.argsort(array_a)
print("Indices of sorted array:", sorted_indices)

# iii. 4 smallest elements
smallest_elements = np.sort(array_a)[:4]
print("4 smallest elements:", smallest_elements)

# iv. 5 largest elements
largest_elements = np.sort(array_a)[-5:]
print("5 largest elements:", largest_elements)

# Part (b): Operations on array with floats
array_b = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])

# i. Integer elements only
integer_elements = array_b[array_b == array_b.astype(int)]
print("\nInteger elements only:", integer_elements)

# ii. Float elements only
float_elements = array_b[array_b != array_b.astype(int)]
print("Float elements only:", float_elements)
```

```python
import numpy as np
#a)
name = input('enter your full name: ')
#divide name into two based on space
space_index = name.find(' ')
if space_index != -1:
    first_name = name[:space_index]
    last_name = name[space_index+1:]
first_name_initial= first_name[0]
last_name_initial= last_name[0]
X= ord(first_name_initial)+ord(last_name_initial)
sales= np.linspace(X,X+200,5)
print(sales)

#b)
taxrate = ((X % 5)+5)/100
adjusted_sales= sales*taxrate
print(adjusted_sales)

#c)
discounted_sales = np.where(sales < X + 100, sales * 0.95, sales * 0.90)
print(discounted_sales)

#d)
expanded_sales = np.tile(sales,(3,1))
print(expanded_sales)
increased_sales = expanded_sales*0.2 + expanded_sales
print(increased_sales)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)

plt.plot(x, np.square(x), label='y = x^2', color='blue')
plt.plot(x, np.sin(x), label='y = sin(x)', color='green')
plt.plot(x, np.exp(x), label='y = e^x', color='red')
plt.plot(x, np.log(np.abs(x) + 1), label='y = log(|x|+1)', color='orange')

plt.ylim(-5, 50)  # Limit y-axis to -5 to 50 for a better scale
plt.title('Multiple Functions on One Graph')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```
