```python
import numpy as np
import matplotlib.pyplot as plt

M= float(input('enter value of M: '))
x= np.linspace(-10,10,100)
y1= M*x**2
y2= M*np.sin(x)

plt.plot(x,y1,'r--',label='y=M*x^2')
plt.plot(x,y2,'b-',label='y=M*sin(x)')
plt.legend()
plt.grid()
plt.title('mathematical function')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Create the dataset
data = {'Subject': ['Math', 'English', 'Science', 'History', 'Art'],
        'Score': [85, 92, 78, 88, 95]}

# 2. Convert to Pandas DataFrame
df = pd.DataFrame(data)

# 3. Plot the scores using Seaborn
plt.figure(figsize=(10, 6))  # Adjust figure size for better readability
ax = sns.barplot(x='Subject', y='Score', data=df, palette='viridis')  # Use a color palette

# Add annotations on top of each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')  # Adjust text position

plt.title('Subject Scores', fontsize=16)
plt.xlabel('Subject', fontsize=12)
plt.ylabel('Score', fontsize=12)10
plt.grid(axis='y', linestyle='--')
plt.ylim(0, 100)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


roll_number = int(input("Enter your roll number: "))
np.random.seed(roll_number)

data = np.random.randn(50)
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes[0, 0].plot(np.cumsum(data), marker='o', linestyle='-')
axes[0, 0].set_title("Cumulative Sum Line Plot")
axes[0, 0].set_xlabel("Index")
axes[0, 0].set_ylabel("Cumulative Sum")
axes[0, 0].grid(True)

noise = np.random.randn(50)
axes[0, 1].scatter(range(50), data + noise, color='r', alpha=0.6)
axes[0, 1].set_title("Scatter Plot with Random Noise")
axes[0, 1].set_xlabel("Index")
axes[0, 1].set_ylabel("Value with Noise")
axes[0, 1].grid(True)

axes[1, 0].axis("off")
axes[1, 1].axis("off")

plt.tight_layout()
plt.show()
```

```python
from google.colab import files
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Upload file
uploaded = files.upload()

# Read the uploaded CSV file
filename = list(uploaded.keys())[0]  # Get the uploaded file name
df = pd.read_csv(filename)  # Read it into a DataFrame

# Line plot for total profit over months
plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x='month_number', y='total_profit', marker='o', color='b')
plt.title('Total Profit Over Months')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.grid()
plt.show()

# Multi-line plot for product sales
plt.figure(figsize=(10, 5))
for column in df.columns[1:-1]:  # Excluding 'month_number' and 'total_profit'
    sns.lineplot(data=df, x='month_number', y=column, label=column)
plt.title('Product Sales Over Months')
plt.xlabel('Month')
plt.ylabel('Units Sold')
plt.legend()
plt.grid()
plt.show()

# Bar chart for all features
plt.figure(figsize=(12, 6))
df.set_index('month_number').plot(kind='bar', figsize=(12, 6))
plt.title('Monthly Sales Data')
plt.xlabel('Month')
plt.ylabel('Values')
plt.legend()
plt.grid()
plt.show()
```
